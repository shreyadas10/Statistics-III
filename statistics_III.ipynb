1. What is a random variable in probability theory ?

-> A random variable is a function that assigns a numerical value to each outcome of a random experiment. The value of the random variable is uncertain until the experiment is performed, and it is determined by the outcome of the experiment.


2. What are the types of random variables?

->  There are several types of random variables, which can be classified based on their properties and characteristics. Here are some of the main types of random variables:

- Discrete Random Variable: A discrete random variable can take on a countable number of distinct values. Examples include:
    - Number of heads in a coin toss
    - Number of students in a class
    - Number of defective products in a batch
- Continuous Random Variable: A continuous random variable can take on any value within a given interval or range. Examples include:
    - Height of a person
    - Temperature of a room
    - Time it takes to complete a task
- Mixed Random Variable: A mixed random variable is a combination of discrete and continuous random variables. Examples include:
    - Time until a certain event occurs (discrete) and the amount of time spent waiting (continuous)
- Categorical Random Variable: A categorical random variable can take on a set of distinct categories or labels. Examples include:
    - Color of a car (red, blue, green, etc.)
    - Type of music (rock, pop, jazz, etc.)
- Ordinal Random Variable: An ordinal random variable can take on a set of distinct categories or labels that have a natural order or ranking. Examples include:
    - Level of education (high school, college, graduate degree, etc.)
    - Level of satisfaction (very satisfied, somewhat satisfied, neutral, etc.)
-. Count Random Variable: A count random variable can take on non-negative integer values that represent the number of occurrences of an event. Examples include:
    - Number of accidents in a given period
    - Number of defects in a product
- Bernoulli Random Variable: A Bernoulli random variable can take on only two possible values (0 and 1) and represents the outcome of a single trial or experiment. Examples include:
    - Coin toss (heads or tails)
    - Quality control inspection (pass or fail)

3. What is the difference between discrete and continuous distributions ?

-> the difference between  discrete and continuous distributions are:

- Number of values: Discrete distributions have a countable number of distinct values, while continuous distributions have an uncountable number of possible values.

- Probability: In discrete distributions, the probability of each value is positive, while in continuous distributions, the probability of each value is zero.

- Cumulative distribution function (CDF): The CDF of a discrete distribution is a step function, while the CDF of a continuous distribution is a continuous function.

4. What are probability distribution functions (PDF)?

-> Probability Distribution Functions (PDFs) are mathematical functions that describe the probability of a random variable taking on a specific value or range of values.

-> Types of PDFs :-

- Discrete PDFs: Describe the probability distribution of discrete random variables, such as the Bernoulli distribution or the Poisson distribution.
- Continuous PDFs: Describe the probability distribution of continuous random variables, such as the Normal distribution or the Exponential distribution.
- Mixed PDFs: Describe the probability distribution of random variables that have both discrete and continuous components.


5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)) ?

->The key differences between CDFs and PDFs are:

- Interpretation: A PDF describes the probability of a specific value or range of values, while a CDF describes the probability that a value is less than or equal to x.
- Calculation: A PDF is typically calculated as a function of the random variable, while a CDF is calculated as the integral of the PDF.
- Properties: A PDF satisfies the properties of non-negativity and normalization, while a CDF satisfies the properties of non-negativity, monotonicity, and boundedness.

6. What is a discrete uniform distribution ?

-> A discrete uniform distribution is a distribution where the random variable X can take on a finite number of distinct values, say x1, x2, ..., xn and each value xi has an equal probability of occurring, say 1/n.


7. What are the key properties of a Bernoulli distribution ?

-> A Bernoulli distribution models a random experiment with only two possible outcomes, typically labeled "success" (1) or "failure" (0), and is characterized by a single parameter, the probability of success (p), where the probability of failure is 1-p.

8. What is the binomial distribution, and how is it used in probability ?

It's used to calculate probabilities in scenarios with two possible outcomes (success or failure).

9. What is the Poisson distribution and where is it applied ?

-> The Poisson distribution is a discrete probability distribution that models the probability of a certain number of events occurring within a fixed interval of time or space, given that these events occur independently and at a constant average rate. It's used in various fields to analyze and predict events like customer arrivals, call center calls, or rare occurrences.

10. What is a continuous uniform distribution ?

-> A continuous uniform distribution is a probability distribution that has the following properties:

1. Interval: The random variable X can take on any value within a specific interval [a, b], where a and b are constants.
2. Equal probability: The probability of X taking on any value within the interval [a, b] is equal.
3. Continuous: The random variable X can take on any value within the interval [a, b], including fractions and decimals.


11. What are the characteristics of a normal distribution ?

-> A normal distribution, also known as a bell curve, is characterized by being symmetrical, unimodal (one peak), and having the mean, median, and mode all equal.

-> the characteristics of a normal distribution are:

- Symmetry:
A normal distribution is perfectly symmetrical around its center (the mean). The right side of the curve is a mirror image of the left side.

- Bell Shape:
The curve resembles a bell, with the highest point at the center and gradually tapering off on either side.

- Unimodal:
It has only one peak or mode, indicating that the data is clustered around a single value.

- Mean, Median, and Mode Equality:
In a normal distribution, the mean (average), median (middle value), and mode (most frequent value) are all equal and located at the center of the distribution.

- Asymptotic:
The tails of the curve extend infinitely towards the horizontal axis but never touch it.

- Standard Deviation:
The standard deviation determines the spread or dispersion of the data around the mean. A smaller standard deviation means the data is clustered tightly around the mean, while a larger standard deviation indicates wider dispersion.

- Central Limit Theorem:
The distribution of the sum (or average) of a large number of independent, identically distributed variables will be approximately normally distributed, regardless of the underlying distribution.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics ?

It's critical in statistics because it allows us to make inferences about populations using sample data, even when we don't know the population's distribution.

16. How do you calculate a Z-score, and what does it represent ?

Z = (X - μ) / σ

Where:

- X = sample value
- μ = population mean
- σ = population standard deviation

Z-score represents:

- Number of standard deviations from the mean
- How unusual a value is in a normal distribution

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_dist(p,k):
    '''parameters :-
       p = probablity of success (0 <= p <=1)
       k = number of trial (0 or 1)
    '''
    if k not in [0,1]:
        raise ValueError("k must be 0 or 1")
    if not 0 <= p <=1:
        raise ValueError("p must be between 0 to 1")

    return  p*k +(1-p)*(1-k)

p = 0.8
k = 1

try:
  print(bernoulli_dist(p,k))
except ValueError as e:
    print(e)
```

9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores


np.random.seed(0)
data = np.random.normal(0, 1, 1000)

z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, density=True)
plt.title('Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python

np.random.seed(0)

num_samples = 1000
sample_size = 100

samples = np.random.exponential(scale=1, size=(num_samples, sample_size))

means = np.mean(samples, axis=1)

plt.hist(means, bins=30, density=True)

x = np.linspace(means.min(), means.max(), 100)
y = np.exp(-((x - means.mean()) / means.std())**2 / 2) / (means.std() * np.sqrt(2 * np.pi))
plt.plot(x, y, 'r--')

plt.title('Central Limit Theorem (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.show()
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset

```python
from scipy import stats
data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 26, 27])

null_hypothesis = "The population mean is equal to 20."
alternative_hypothesis = "The population mean is not equal to 20."

alpha = 0.05

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

population_std = 5

z_statistic = (sample_mean - 20) / (population_std / np.sqrt(len(data)))

p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

print("Null Hypothesis:", null_hypothesis)
print("Alternative Hypothesis:", alternative_hypothesis)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-statistic:", z_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python

data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 26, 27])

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)


confidence_level = 0.95

standard_error = sample_std / np.sqrt(len(data))

degrees_of_freedom = len(data) - 1
critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

margin_of_error = critical_value * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Confidence Level:", confidence_level * 100, "%")
print("Confidence Interval: (", lower_bound, ", ", upper_bound, ")")

print("We are", confidence_level * 100, "% confident that the true population mean lies between", lower_bound, "and", upper_bound)
```

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution

```python
def calculate_skewness_kurtosis(data):

    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    return skewness, kurtosis

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

skewness, kurtosis = calculate_skewness_kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

plt.hist(data, bins=30, density=True)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
